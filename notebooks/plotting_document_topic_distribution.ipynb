```python
def generate_coordinates(data_size, number_of_rows, row_length, column_length):
    single_scan_size = (data_size + 1) / number_of_rows
    column_size = int(math.floor(single_scan_size \
        * (column_length / (row_length + column_length))))
    row_size = single_scan_size - column_size 
    
    x_coord = 0
    y_coord = 7
    row_scan = True
    scanning_right = True
    coordinates = {}
    for j in range(0, number_of_rows):
        for i in range(0, row_size):
            doc_id = i + column_size * j + row_size * j
            coordinates[doc_id] = {'x': x_coord, 'y': y_coord}
            # Switching horizontal scanning to vertical.
            if (i % (row_size - 1) == 0 and i != 0):
                y_coord -= 1
                continue
            if scanning_right:
                x_coord += 1
            else:
                x_coord -= 1         
        # Reversing the scanner's direction.
        scanning_right = not scanning_right
    return coordinates
```

```python
import math

coordinates_mapping = generate_coordinates(6327, 8, 62, 1.25)
```

```python
import numpy as np

probability_matrix = np.zeros(shape=(8,776))
topic_id = 7

i = 0
for doc in document_topic_probabilities:
    doc_id = inv_map[i]
    if doc_id in coordinates_mapping:
        doc_coord = coordinates_mapping[doc_id]
        probability_matrix[doc_coord['y']][doc_coord['x']] = doc[topic_id]
    i += 1
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

x_vector = np.zeros(shape=(776, 1))
y_vector = np.zeros(shape=(8, 1))

for i in range(0, 776):
    x_vector[i] = i
    
for i in range(0, 8):
    y_vector[i] = i

layout = go.Layout(
    title='Topic distribution in a corpus.',
    height=350
)
    
data = [
    go.Heatmap(
        z=probability_matrix,
        x=x_vector,
        y=y_vector,
        colorscale='Electric'
    )
]

fig = go.Figure(data=data, layout=layout)

py.iplot(fig)
```
